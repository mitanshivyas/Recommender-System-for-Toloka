### Imp point:

* From here onwards, we have two ways in which we can incorporate other features into our model:



1.   Either we cluster similar features together just like how we did in terms of title and description(the most similar titles and descriptions to our query would be recommended). For example: If a worker has been selecting tasks that has a certain amount of pay and grade, then we could cluster all the tasks provide similar amount of pay and grade and other features and then show our recommendations. However, that could have certain drawbacks, a worker who has been performing tasks with comparaively less pay would most likely not be able get tasks with high pay recommended. Which could be of a disadvantage for the workers.
2.   The second way is to cluster tasks based on simply similarity in title and description and then to use other features (pay, grade, trustworthiness) as a way to sort those tasks. So the top most recommended tasks would be similar to each other, as well as the worker would be provided with the selection of tasks that have best pay, grade and other features.

### Function that builds recommendations automatically

```python
def build_recommendations(dataframe, number_of_recs):
  user_ids = dataframe['user_id'].unique()
  corpus = df2['data']
  titles = df2['title']
  titles = titles.tolist() 
  corpus = corpus.tolist()
  corpus_embedding = model.encode(corpus, convert_to_tensor=True)
  top_k = min(number_of_recs, len(corpus))

  for i in range(len(user_ids)):
    new_df = dataframe.loc[dataframe['user_id'] == user_ids[i]]
    new_df = new_df.reset_index(drop=True)
    query = ' '.join(new_df.title)

    query_embedding = model.encode(query, convert_to_tensor=True)
    cos_scores = util.cos_sim(query_embedding, corpus_embedding)[0]
    top_results = torch.topk(cos_scores, k=top_k)
    print("---------------------------")
    print("User ID:", user_ids[i])
    print("---------------------------")
    # indices = new_df.index[new_df['user_id'] == i].tolist()
    # j = indices[0]
    print("User Language: " + str(new_df['user_lang'][0]))
    for score, idx in zip(top_results[0], top_results[1]):
      if df2['task_language'][int(idx)] in df2['user_lang'][int(idx)]:
        print("Task Language: " + str(df2['task_language'][int(idx)]))
        print(f'{round(score.item(), 3)} | {titles[idx]}')
      else:
        pass
```
