```python
import numpy as np

def map_ndcg_k(recommended_titles, ground_truth_titles, k):
    """
    Calculates the MAP@k and NDCG@k scores for a recommender system.
    
    Args:
    recommended_titles (list): A list of recommended task titles.
    ground_truth_titles (list): A list of ground truth task titles.
    k (int): The maximum number of recommendations to consider.
    
    Returns:
    float: The MAP@k score.
    float: The NDCG@k score.
    """
    
    # Get the top-k recommended titles
    recommended_titles = recommended_titles[:k]
    
    # Calculate the precision at each position
    precision = []
    for i in range(k):
        if recommended_titles[i] in ground_truth_titles:
            precision.append(1)
        else:
            precision.append(0)
    
    # Calculate the average precision
    ap = np.mean(precision)
    
    # Calculate the ideal DCG
    ideal_dcg = 0
    for i in range(min(k, len(ground_truth_titles))):
        ideal_dcg += 1 / np.log2(i+2)
    
    # Calculate the DCG
    dcg = 0
    for i in range(min(k, len(recommended_titles))):
        if recommended_titles[i] in ground_truth_titles:
            dcg += 1 / np.log2(i+2)
    
    # Calculate the NDCG
    ndcg = dcg / ideal_dcg
    
    return ap, ndcg
```

```python
recommended_titles = ['Does the provided link contain coupons?', 'Microsoft Bing App test','Is the product relevant to the query?' , 'Does the store have this product?', 'MMWatermarkImageDetection']
```

```python
project_ids = set(income_data[income_data['uid'] == 1430094402]['project_id'])
# Filter the task data to include only the tasks for the specified projects
project_tasks = task_data[task_data['project_id'].isin(project_ids)]

# Get a list of the titles for the project tasks
ground_truth_titles = list(project_tasks['title'])

# Print the list of project titles
print(ground_truth_titles)
```

```python
ap, ndcg = map_ndcg_k(recommended_titles, ground_truth_titles, 5)

for i in recommended_titles:
  print(i)
print("----------------------------")
print(f'MAP@{5} score: {ap:.2f}')
print(f'NDCG@{5} score: {ndcg:.2f}')
```
